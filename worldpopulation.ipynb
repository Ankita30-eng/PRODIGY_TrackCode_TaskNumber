```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("worldpopulationdata.csv")

# Filter datasets
male_df = df[df["Series Name"] == "Population, male"]
female_df = df[df["Series Name"] == "Population, female"]

# Merge on 'Country Code' to get male and female in same row
merged_df = pd.merge(male_df[['Country Code', '2022']],
                     female_df[['Country Code', '2022']],
                     on='Country Code',
                     suffixes=('_male', '_female'))

# Calculate total population
merged_df["Total"] = merged_df["2022_male"] + merged_df["2022_female"]

# Get top 10 countries by total population
top_10 = merged_df.sort_values(by="Total", ascending=False).head(10)

# Plot stacked bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Code", y="2022_female", data=top_10, color="red", label="Female Population")
sns.barplot(x="Country Code", y="2022_male", data=top_10, color="green", label="Male Population",
            bottom=top_10["2022_female"])

plt.title("Top 10 Countries by Gender Population in 2022")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("worldpopulationdata.csv")

# Filter male and female population data
male_df = df[df["Series Name"] == "Population, male"]
female_df = df[df["Series Name"] == "Population, female"]

# Merge male and female data
merged_df = pd.merge(
    male_df[["Country Code", "Country Name", "2022"]],
    female_df[["Country Code", "2022"]],
    on="Country Code",
    suffixes=("_male", "_female")
)

# Calculate total population
merged_df["Total"] = merged_df["2022_male"] + merged_df["2022_female"]

# Get bottom 10 countries by total population
bottom_10_countries = merged_df.sort_values(by="Total", ascending=True).head(10)

# Plot stacked bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Code", y="2022_female", data=bottom_10_countries, color="red", label="Female Population")
sns.barplot(x="Country Code", y="2022_male", data=bottom_10_countries, color="green", label="Male Population",
            bottom=bottom_10_countries["2022_female"])
plt.title("Bottom 10 Countries by Gender Population in 2022")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
